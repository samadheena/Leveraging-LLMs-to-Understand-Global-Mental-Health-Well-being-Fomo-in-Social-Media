```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftConfig, PeftModel

base_model = "mistralai/Mistral-7B-Instruct-v0.2"
adapter = "GRMenon/mental-health-mistral-7b-instructv0.2-finetuned-V2"

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(
    base_model,
    add_bos_token=True,
    trust_remote_code=True,
    padding_side='left'
)

# Create peft model using base_model and finetuned adapter
config = PeftConfig.from_pretrained(adapter)
model = AutoModelForCausalLM.from_pretrained(config.base_model_name_or_path,
                                             load_in_4bit=True,
                                             device_map='auto',
                                             torch_dtype='auto')
model = PeftModel.from_pretrained(model, adapter)

device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
model.eval()


```

```python

prompt = """  Act as a classification bot and classify the user's statement into one out of the 3 categories: Social Media Addiction or Social Isolation or Cyberbullying
You will only respond with the category. Do not include the word "Category". Do not provide explanations or notes.

Here are the key phrases to look for in order to identify the category:

Social Media Addiction: always rely on social media to improve mood, often rely on social media to improve mood, often feel the urge to check social media, almost constantly feel the urge to check social media, social media use often interfered with work, studies, or chores, moderately anxious when cannot access social media, very anxious when cannot access social media

Social Isolation: always feel more connected to people online than in real life, often feel more connected to people online than in real life, participation in offline social activities has significantly decreased, often feel lonely even if I am active on social media, always feel lonely even if I am active on social media, Social media is exclusively my primary way of socializing, Social media is mostly my primary way of socializing, much more likely to share personal matters online than with people offline

Cyberbullying: often encounter negative comments, harassment, or bullying on social media, almost constantly encounter negative comments, harassment, or bullying on social media, experienced impersonation or identity theft as forms of cyberbullying, experienced Mean or hurtful comments as forms of cyberbullying, experienced Threats or intimidation as forms of cyberbullying, have experienced
exclusion from online groups as forms of cyberbullying, have experienced Spreading of rumors or lies as forms of cyberbullying, Negative interactions on social media make me feel extremely upset, Negative interactions on social media make me feel very upset, slightly upset by negative interactions on social media, moderately upset by negative interactions on social media, very upset by negative interactions on social media, extremely upset by negative interactions on social media, Cyberbullying has significantly lowered my self-esteem or self-image

Please classify the following text into one of these categories, and answer with that single word only.

[INST] user: I spend 3-4 hours on social media every day.
I often feel the urge to check social media.
My social media use rarely interferes with my work, studies, or chores.
I often rely on social media to improve my mood.
I feel very anxious when I cannot access social media.
I sometimes encounter negative comments, harassment, or bullying on social media.
I have not experienced any form of cyberbullying.
Negative interactions on social media leave me unaffected.
I typically report cyberbullying to the platform.
Cyberbullying has had no effect on my self-esteem or self-image.
I sometimes feel more connected to people online than in real life.
My participation in offline social activities has remained the same since using social media.
I sometimes feel lonely, even if I am active on social media.
Social media is equally my primary way of socializing online and offline.
I am about the same likely to share personal matters online rather than with people I know offline.[/INST]"""

inputs = tokenizer(prompt, return_tensors="pt")

# generate output
outputs = model.generate(input_ids=inputs["input_ids"].to("cuda"),max_new_tokens=8)

print(tokenizer.batch_decode(outputs)[0])
```
